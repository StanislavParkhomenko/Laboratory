```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./1_8.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show();
```

```python
bin_img = img.copy()
threshold = 10
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val
```

```python
plt.figure(figsize=(10,10))
plt.imshow(bin_img)
plt.show();
```

```python
kernel = np.ones((2,2),np.uint8)
```

```python
erosion = cv2.erode(bin_img,kernel,iterations = 1)

plt.figure(figsize=(10,10))
plt.imshow(erosion, cmap='gray')
plt.show();
```

```python
dilation = cv2.dilate(bin_img,kernel,iterations =1)

plt.figure(figsize=(10,10))
plt.imshow(dilation, cmap='gray')
plt.show();
```

```python
img = cv2.imread("street.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(img)
plt.show();
```

```python
bin_img = img.copy()
threshold = 170
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val
```

```python
plt.figure(figsize=(20,20))
plt.imshow(bin_img)
plt.show();
```

```python
kernel = np.ones((2,2),np.uint8)
```

```python
erosion = cv2.erode(bin_img,kernel,iterations = 1)

plt.figure(figsize=(20,20))
plt.imshow(erosion, cmap='gray')
plt.show();
```

```python
opening = cv2.morphologyEx(bin_img,cv2.MORPH_GRADIENT,kernel)

plt.figure(figsize=(20,20))
plt.imshow(opening, cmap='gray')
plt.show();
```
